```python
from helpers import *

returns_assets = pd.read_excel('data/BloombergReturnsDaily.xlsx', index_col=0, parse_dates=True)
#returns_benchmark = pd.read_excel('data/DailyReturnsBenchmark.xlsx', index_col=0, parse_dates=True)
# calculate returns benchmark as equal weights of the returns of the assets
returns_benchmark = returns_assets.mean(axis=1) 
# make the returns benchmark a dataframe witht he same index as assets and name the column of the average return SP500
returns_benchmark = pd.DataFrame(returns_benchmark, index=returns_assets.index, columns=['SP500'])



window_size_months = 5 
start_index = window_size_months + 1  
date_list = []

months = returns_assets.index.to_period('M').unique()

for month in months[start_index:]:

    current_date = month.start_time
    date_list.append(current_date)

    start_date = month.start_time - pd.DateOffset(months=window_size_months+1)
    end_date = month.start_time - pd.DateOffset(months=1)

    assets_range = returns_assets.loc[start_date:end_date].copy()
    bench_range = returns_benchmark.loc[start_date:end_date].copy()

    assets_uniform = transform_to_uniform(assets_range)
    bench_uniform = transform_to_uniform(bench_range)

    bench_uniform.to_csv(f"data_uniform/bloom_benchmark_{current_date.strftime('%Y-%m-%d')}.csv")
    assets_uniform.to_csv(f"data_uniform/bloom_sp100_{current_date.strftime('%Y-%m-%d')}.csv")

    print(f"Current Date: {current_date}, Start Date: {start_date}, End Date: {end_date}")

date_list = pd.Series(date_list)
date_list.to_csv('data_uniform/dates_bloom.csv', index=False)
```
